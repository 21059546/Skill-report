```python
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot
import pandas as pd
import csv
```

```python
df = pd.read_csv( 'POI.csv')
x = df
x_np = np.array(x)  

n_clusters = 7                
estimator = KMeans(n_clusters) 

estimator.fit(x)

markers = ['*', 'v', '+', '^', 's','o','x']      # List of markup styles
colors = ['r', 'g', 'm', 'c', 'orange','b','pink']  
labels = estimator.labels_      # Get cluster labels
centers = estimator.cluster_centers_

pyplot.figure(figsize=(9, 6))
pyplot.title("Clustering of POI in La defense ", fontsize=25)
pyplot.xlabel('Longitude', fontsize=18)
pyplot.ylabel('Latitude', fontsize=18)

pyplot.scatter(centers[:,1],centers[:,0],c = 'grey',s=200,marker = 'o')

for i in range(n_clusters):    # Iterate over all coordinate points and draw scatter plots
    members = labels == i      # Members is a Boolean array  
    pyplot.scatter(
        x_np[members, 1],      
        x_np[members, 0],      
        marker = markers[i],   
        c = colors[i]          # Mark color
    )   

pyplot.grid()
pyplot.show()
```
